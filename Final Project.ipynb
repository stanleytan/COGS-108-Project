```python
%matplotlib inline

#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#configure libaries
#seaborn library
sns.set()
sns.set_context('talk')

#don't display too many rows/colms
pd.options.display.max_rows = 7

#round decimals when displaying DataFrames
pd.set_option('precision', 2)

#set palette to colorblind
palette = sns.color_palette("colorblind")
sns.set_palette(palette)
```

```python
df_reviews = pd.read_csv("employee_reviews.csv")
```

```python
df_review_ratings = df_reviews[['overall-ratings','work-balance-stars','culture-values-stars','carrer-opportunities-stars','comp-benefit-stars','senior-mangemnet-stars']]
df_review_ratings
```

```python
cols = ['overall-ratings', 'work-balance-stars', 'culture-values-stars',
       'carrer-opportunities-stars', 'comp-benefit-stars',
       'senior-mangemnet-stars']
df_review_ratings[cols]=df_review_ratings.apply(pd.to_numeric, errors='coerce', axis=1)
df_review_ratings.columns = ['overall', 'work balance', 'culture values', 'career opportunities', 'benefits', 'senior management']
```

```python
plt.figure(figsize=(15,10))
#plt.xticks(rotation=45)
overall = sns.boxplot(data=df_review_ratings)
overall.set_title('Overall Ratings', fontsize=25)
```
